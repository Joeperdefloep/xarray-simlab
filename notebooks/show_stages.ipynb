```python
import xsimlab as xs
```

```python
@xs.process
class In1:
    v = xs.variable(intent='in')

@xs.process
class In2:
    v  = xs.foreign(In1,"v")
    
@xs.process
class In3:
    v = xs.foreign(In1,"v")
    
@xs.process
class Inout2:
    v = xs.foreign(In1, 'v',intent='inout')

@xs.process
class Inout3:
    v = xs.foreign(In1, 'v', intent='inout')
```

```python
model = xs.Model(
                {"in2":In2,"in1": In1, "io2": Inout2, "io3":Inout3,"in3":In3},
                custom_dependencies={"io3":{"in3","io2"},"io2":{"in2","in1"}},
                strict_order_check=False,
            )#.visualize()
model.visualize(show_feedbacks=False)
```

```python
model.visualize(show_feedbacks=False)
```

```python
model.visualize(show_variable_stages=("in1","v"))
```
